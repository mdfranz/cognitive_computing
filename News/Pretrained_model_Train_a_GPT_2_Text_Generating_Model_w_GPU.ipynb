## Generate Text From The Pretrained Model

If you want to generate text from the pretrained model, not a finetuned model, pass `model_name` to `gpt2.load_gpt2()` and `gpt2.generate()`.

This is currently the only way to generate text from the 774M or 1558M models with this notebook.

```
model_name_un_tr = "355M"

gpt2.download_gpt2(model_name=model_name_un_tr)
```

```
sess = gpt2.start_tf_sess()

gpt2.load_gpt2(sess, model_name=model_name_un_tr)
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
path = "/content/drive/Shared drives/Cognitive Computing Project /news_clean.csv"
df = pd.read_csv(path)
```

```
df
```

```
df['GenGPT2355M'] = 0
for i in range(0,len(df)):
  x = df['First'][i]
  x = """%s""" % (x)
  blah = gpt2.generate(sess,
              model_name=model_name_un_tr,
              prefix=x,
              length=450,
              temperature=0.7,
              top_p=0.9,
              nsamples=1,
              batch_size=1,
              return_as_list = True
              )
  df['GenGPT2355M'][i] = blah[0]
  print(i)
```
